```python
import cv2
import numpy as np

# Create a VideoCapture object
cap = cv2.VideoCapture(0)

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Failed to open camera")
    exit()

# Define the color filters
color_filters = [
    (0, 0, 255),   # Red
    (0, 255, 0),   # Green
    (255, 0, 0),   # Blue
    (255, 255, 0)  # Yellow
]

while True:
    # Read frame from the camera
    ret, frame = cap.read()

    if not ret:
        print("Failed to read frame")
        break

    # Split the frame into four parts
    height, width, _ = frame.shape
    part_height = height // 2
    part_width = width // 2

    # Apply color filters to each part of the frame
    for i, color in enumerate(color_filters):
        x_start = i % 2 * part_width
        y_start = i // 2 * part_height
        x_end = x_start + part_width
        y_end = y_start + part_height

        # Apply color filter to the part of the frame
        frame[y_start:y_end, x_start:x_end] = np.array(color, dtype=np.uint8)

    # Display the frame
    cv2.imshow("Video", frame)

    # Check if 'q' is pressed to exit the loop
    if cv2.waitKey(1) == ord('q'):
        break

# Release the VideoCapture object and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Function to crop and display video stream
def crop_video():
    # Open the camera
    cap = cv2.VideoCapture(0)  # 0 represents the default camera

    # Check if the camera is opened successfully
    if not cap.isOpened():
        print("Failed to open the camera")
        return

    while True:
        # Capture a frame from the camera
        ret, frame = cap.read()

        # Check if the frame is captured successfully
        if not ret:
            print("Failed to capture frame")
            break

        # Specify the region of interest (ROI) coordinates
        x, y, w, h = 100, 100, 300, 200  # Adjust these values according to your desired ROI

        # Crop the frame to the specified ROI
        cropped_frame = frame[y:y+h, x:x+w]

        # Display the cropped video
        cv2.imshow("Cropped Video", cropped_frame)

        # Check for the 'q' key press to exit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the camera and close all windows
    cap.release()
    cv2.destroyAllWindows()

# Call the function to start capturing and displaying the cropped video stream
crop_video()
```
